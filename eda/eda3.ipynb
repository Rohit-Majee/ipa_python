## Tips dataset EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("tips.csv")
df.head()
```

```python
plt.scatter("total_bill","tip", data = df, label = 'tips', color="green")
plt.show()
```

```python
plt.style.use("dark_background")
```

```python
plt.style.use('seaborn-v0_8-pastel')
plt.figure(figsize= (6, 6))

plt.subplot(3, 3, 1)
plt.scatter(df['total_bill'], df['tip'], marker='.', color='orange')
plt.title("Total bill vs Tip")
plt.xlabel("Total bills")
plt.ylabel("Tip")

plt.subplot(3, 3, 3)
plt.pie(df.day.value_counts(), labels = df.day.value_counts().index, autopct = "%.2f%%")
plt.title("Day wise tip")

plt.subplot(3, 3, 5)
plt.bar(x = df.day.value_counts().index, height = df.day.value_counts())
plt.title("Day vs Tip")
plt.xlabel("Day")
plt.ylabel("Tip")

plt.subplot(3, 3, 7)
plt.bar(x = df['total_bill'], height = df['tip'])
plt.title("Total bill vs Tip")
plt.xlabel("Total bills")
plt.ylabel("Tip")

plt.subplot(3, 3, 9)
plt.plot(df['total_bill'], df['tip'])
plt.title("Total bill vs Tip")
plt.xlabel("Total bills")
plt.ylabel("Tip")

plt.show()
```
