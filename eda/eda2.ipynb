## **Matplotlib**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1,50, 10)
y = np.random.randint(1,50, 10)
print(x,y ,sep='\n')
```

```python
plt.plot(x,y )
```

```python
plt.plot(x,y , color ="red", linestyle = 'dotted')
```

```python
plt.plot(x,y , color ="red")
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y , color ="red",linestyle = 'dashed')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y , color ="red",linestyle = 'dashed', alpha = 0.3)
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y , color ="red",linestyle = 'dashed', alpha = 0.9,
         marker = "o")
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y , color ="red",linestyle = 'dashed', alpha = 0.9,
         marker = "o", markerfacecolor= "b")
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y , color ="red",linestyle = 'dashed', alpha = 0.9,
         marker = "*", markerfacecolor= "b", markersize = 10)
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y , color ="red",linestyle = 'dashed', alpha = 0.9,
         marker = "*", markerfacecolor= "blue", markersize = 10, markeredgecolor = 'black')
plt.show()
```

```python
plt.title("Random chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y , color ="black",linestyle = 'dashed', alpha = 0.9,
         marker = "*", markerfacecolor= "blue", markersize = 10, markeredgecolor = 'black',
         linewidth = 5)
plt.show()
```

```python
x = np.arange(1,11)
y = np.sin(x)
z = np.cos(x)
print(x,y,z, sep="\n")
```

```python
plt.plot(x,y,label= "SINQ")
plt.show()
```

```python
plt.plot(x,z,label= "COSQ", color='red')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.xticks(x)
plt.yticks(y)
plt.yticks(z)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,z,label= "COSQ", color='red')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.xticks(x)
plt.yticks(y)
plt.yticks(z)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(x,z,label= "COSQ", color='red')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values") 
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,z,label= "COSQ", color='green')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(3,3,1)
plt.plot(x,z,label= "COSQ", color='red')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values") 
plt.legend()

plt.subplot(3,3,5)
plt.plot(x,z,label= "COSQ", color='purple')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()


plt.subplot(3,3,9)
plt.plot(x,z,label= "COSQ", color='green')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(x,z,label= "COSQ", color='red')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values") 
plt.legend(loc="upper left")

plt.subplot(1,2,2)
plt.plot(x,z,label= "COSQ", color='green')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend(loc="lower left")

plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.subplot(1,2,1)
plt.plot(x,z,label= "COSQ", color='red')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values") 
plt.legend(loc="upper left")

plt.subplot(1,2,2)
plt.plot(x,z,label= "COSQ", color='green')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend(loc="lower left")
plt.subplots_adjust(wspace=0.5)

plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(x,z,label= "COSQ", color='red')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values") 
plt.legend(loc="upper left")
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,z,label= "COSQ", color='green')
plt.plot(x,y,label= "SINQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend(loc="lower left")
plt.grid()


plt.show()
```
