### merge and concat

```python
data1 = {"Name":['A', 'B', 'C'],
         'age' : [23,45,67],
         "salary" : [10000,20000,30000]}

data2 = {"Name":['A', 'BB', 'C'],
         'age' : [23,54,67],
         "salary" : [10000,25000,30000]}
```

```python
d1 = pd.DataFrame(data1)
d2 = pd.DataFrame(data2)
```

```python
d2
```

```python
# pd.concat([d1,d2],ignore_index=True, axis=1)
```

```python
d12 = pd.concat([d1,d2],ignore_index=True)
d12.head()
```

```python
d12.drop_duplicates()
```

### **Merge**

```python
pd.merge(d1,d2,on='age',how='outer')
```
