```python
#1
l1 = eval(input("Enter the list: "))

if l1[:2] == l1[-2:]:
    print("Valid slice")
else:
    print("Invalid slice")
```

```python
#2
s1 = input("Enter the string: ")
if s1 == s1[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
    
```

```python
#3
s2 = input("Enter the string: ")
if s2[-1].isdigit():
    print(" Ends with digit")
else:
    print("Ends with character")
```

```python
#4
s3 =  input("Enter the string: ")

if s3[0].isupper():
    print("Starts with uppercase")
elif s3[-1].islower():
    print("Ends with lowercase")
else:
    print("Other case")
```

```python
#5
l2 = eval(input("Enter the list: "))

mid = len(l2) // 2
left = sum(l2[:mid])
right = sum(l2[mid:])

if left > right:
    print("First half greater")
elif left == right:
    print("Equal halves")
else:
    print("Second half greater")
```

```python
#6
s4 =  input("Enter the string: ")
if len(s4) >= 5:
    mid = len(s4) // 2
    if s4[mid].lower() in ('a','e','i','o','u'): 
        print("Middle character is a vowel")
    else:
        print("Middle character is not a vowel")
else:
    print("String too short")
```

```python
#7
s5 =  input("Enter the string: ")

if s5[0] == "A":          
    if s5[-1] == "Z":     
        print("Valid AZ string")
    else:
        print("Starts with A but invalid end")
else:
    print("Invalid string")
```

```python
#8
s6 =  input("Enter the string: ")

if s6[:2].isdigit():
    if s6[-2:].isalpha():
        print("valid format")
    else:
        print("invalid format")
else:
    print("invalid format")
```

```python
#9
l2 = eval(input("Enter the list: "))

if len(l2) % 2 == 0:
    if l2[:2] == l2[-2:]:
        print("Slices are same")
    else:
        print("Slices are different")
else:
    print("Odd length list")
```
