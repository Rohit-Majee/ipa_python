```python
## Bank question
class BankAccount(ABC):
    def __init__(self, customer_name, date_of_birth, account_number, balance=0):
        self.customer_name = customer_name        
        self.__date_of_birth = date_of_birth      
        self.__account_number = account_number    
        self.__balance = balance                

    def get_balance(self):
        return self.__balance

    def get_account_number(self):
        return self.__account_number

    def display_account_details(self):
        print("\n--- Account Details ---")
        print(f"Customer Name : {self.customer_name}")
        print(f"Account Number: {self.__account_number}")
        print(f"Balance       : {self.__balance}")

    def _update_balance(self, amount):
        self.__balance += amount

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass
```

```python
class SavingsAccount(BankAccount):
    minimum_balance = 1000

    def __init__(self, customer_name, date_of_birth, account_number, balance=0):
        super().__init__(customer_name, date_of_birth, account_number, balance)

    def deposit(self, amount):
        if amount > 0:
            self._update_balance(amount)
            print(f"Deposited {amount}. New Balance: {self.get_balance()}")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if amount > 0:
            if self.get_balance() - amount >= self.minimum_balance:
                self._update_balance(-amount)
                print(f"Withdrawn {amount}. New Balance: {self.get_balance()}")
            else:
                print("Withdrawal denied! Minimum balance requirement not maintained")
        else:
            print("Withdrawal amount must be positive")

```

```python
class CurrentAccount(BankAccount):
    overdraft_limit = 1000
    def __init__(self, customer_name, date_of_birth, account_number, balance=0):
        super().__init__(customer_name, date_of_birth, account_number, balance)

    def deposit(self, amount):
        if amount > 0:
            self._update_balance(amount)
            print(f"Deposited {amount}. New Balance: {self.get_balance()}")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if amount > 0:
            if self.get_balance() - amount >= -self.overdraft_limit:
                self._update_balance(-amount)
                print(f"Withdrawn {amount}. New Balance: {self.get_balance()}")
            else:
                print("Withdrawal denied! Overdraft limit exceeded.")
        else:
            print("Withdrawal amount must be positive.")

```

```python
savings = SavingsAccount("Alice", "1995-06-15", "SA12345", 2000)
savings.deposit(500)
savings.withdraw(1000)
savings.display_account_details()
```

```python
current = CurrentAccount("Bob", "1990-03-10", "CA67890", 100)
current.withdraw(500)
current.deposit(1000)
current.display_account_details()
```
