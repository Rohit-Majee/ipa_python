```python
#1count the number of words present in the file

with open("a.txt",'r') as f:
    words = f.read().split()
    print(f"total words : {len(words)}")
```

```python
#2 extract the unique words and count them
with open("a.txt",'r') as f:
    words = f.read().split()
    unique_words = set(words)
    print(f"Unique set: {unique_words} \nword count: {len(unique_words)}")
```

```python
#3 create a dictionary of word count
freq = {}
with open("a.txt",'r') as f:
    words = f.read().split()
    for i in words:
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1

    print(f"the word frequency : {freq}")
```

```python
import csv

data = [
    {"name": "Rohit", "age": 21, "branch": "IT"},
    {"name": "Ananya", "age": 22, "branch": "CSE"},
    {"name": "Rahul", "age": 20, "branch": "ECE"}
]

fieldnames = ["name", "age", "branch"]

with open("data.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)

    writer.writeheader()     
    writer.writerows(data)    
```

```python
with open("data.csv", "r") as f:
    reader = csv.DictReader(f)

    for row in reader:
        print(row)
```
