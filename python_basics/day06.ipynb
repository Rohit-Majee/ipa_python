```python
class A:
    def __init__(self,a):
        self.a = a

    def __add__(self,other):
        return self.a + other.a

    def __sub__(self,other):
        return self.a - other.a
    
    def __eq__(self,other):
        return self.a == other.a
    
    

obj1 = A(3)
obj2 = A(4)
print(obj1 + obj2)
print(obj1 - obj2)
print(obj1 == obj2)
```

```python
class Vehicle:
    def __init__(self, vehicle_id, brand):
        self.vehicle_id = vehicle_id
        self.brand = brand

    def calculate_rent(self):
        print("Calculating vehicle rent")



class Car(Vehicle):
    def __init__(self, vehicle_id, brand, price_per_day):
        super().__init__(vehicle_id, brand)
        self.price_per_day = price_per_day

    def calculate_rent(self, days):
        return self.price_per_day * days


class Bike(Vehicle):
    def __init__(self, vehicle_id, brand, price_per_hour):
        super().__init__(vehicle_id, brand)
        self.price_per_hour = price_per_hour

    def calculate_rent(self, hours):
        return self.price_per_hour * hours


class Truck(Vehicle):
    def __init__(self, vehicle_id, brand, price_per_km):
        super().__init__(vehicle_id, brand)
        self.price_per_km = price_per_km

    def calculate_rent(self, distance):
        return self.price_per_km * distance


car = Car("C101", "Toyota", 2000)
bike = Bike("B202", "Yamaha", 100)
truck = Truck("T303", "Tata", 50)
jcb = Truck("JC06", "JCB", 30)

print(f"{car.brand} rent : {car.calculate_rent(3)}")
print(f"{bike.brand} rent : {bike.calculate_rent(5)}")
print(f"{truck.brand} rent : {truck.calculate_rent(3000)}")
print(f"{jcb.brand} rent : {jcb.calculate_rent(300)}")
```
