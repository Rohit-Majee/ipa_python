# list comprehension

```python
#1
l1= [i for i in range(11)]

even_square = [i**2 for i in l1 if i%2 ==0]
even_square
```

```python
#2
s1 = "all vowel in the line IN"
vowel_word = [i for i in s1.split() if i[0].lower() in "aeiou"]
vowel_word
```

```python
#3
l2 = [4,5,5,5,6,7,7,9,9]
unique_list = {i for i in l2 if i>5}
unique_list
```

```python
#4
set1_50 = {i for i in range(1,51) if i%3==0 and i%5==0}
set1_50
```

```python
#5
cubes = {i:i**3 for i in range(1,6)}
cubes
```

```python
#6
students = {'a': 70, 'b' : 45, 'c': 60 , 'd':90}
d1 = {i:students[i] for i in students if students[i]>=60}
d1
```

# class and objects

```python
#1 bank

class Bank:
    bname = "SBI"
    loc = 'West Bengal'
    timing = "10a.m - 4p.m"
    ifsc = 132


    def __init__(self,name, age, phone, pan, balance):
        self.name = name
        self.age = age
        self.phone = phone
        self.pan = pan
        self.balance = balance

    def show_balance(self):
        print(self.balance)

    def withdraw(self, amount):
        if amount > self.balance:
            print("Not enough balance")
            
        else:
            self.balance = self.balance-amount
            print("withdraw successfull!")
            print(f"New Balance : {self.balance}")

            

    def deposit(self,amount):
        self.balance += amount
        print("deposite successfull!!")
        print(f"New Balance : {self.balance}")

    @classmethod
    def change_bname(cls,name):
        cls.bname = name


```

```python
c1 = Bank("aman",22, 65667890, 856789, 9000 )
c2 = Bank("Mohit",27, 56789765, 1234432123, 19000 )
```

```python
c1.deposit(1000)
c1.show_balance()
```

```python
c1.withdraw(1000)
c1.show_balance()
```
