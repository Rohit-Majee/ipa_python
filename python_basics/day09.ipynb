## **Exception handling**

```python
a = int(input("Enter a value for a : ")) 
b = int(input("Enter a value for b : ")) 

c = a/b

print(c)
```

```python
try:
    a = int(input("Enter a value for a : ")) 
    b = int(input("Enter a value for b : ")) 

    c = a/b

    print(c)
except ZeroDivisionError:
    print("Division by zero is not allowed")
```

```python
try:
    a = int(input("Enter a value for a : ")) 
    b = int(input("Enter a value for b : ")) 

    c = a/b

    print(c)
except ZeroDivisionError:
    print("Division by zero is not allowed")
except ValueError:
    print("Please enter only integer value")
```

```python
def safe_div():
    try:
        a = int(input("Enter a value for a : ")) 
        b = int(input("Enter a value for b : ")) 

        c = a/b

        print(c)
    except ZeroDivisionError:
        print("Division by zero is not allowed")
        safe_div()
    except ValueError:
        print("Please enter only integer value")
        safe_div()

safe_div()
```

### generic eexception handling


```python
try:
    a = int(input("Enter a value for a : ")) 
    b = int(input("Enter a value for b : ")) 

    c = a/b

    print(c)
except Exception as e:
    print(f"Exception is handled, the Error : {e}")
```

```python
try:
    a = 100000000000
    for i in range(a):
        print(i,end=' ')
except Exception as e:
    print(f"Exception is handled, the Error : {e}")
# # keyboard interuption is not handled
```

```python
try:
    a = int(input("Enter a value for a : ")) 
    b = int(input("Enter a value for b : ")) 

    c = a/b

    print(c)
except ZeroDivisionError:
    print("Division by zero is not allowed")
except ValueError:
    print("Please enter only integer value")
else:
    print("Division successfull")
finally:
    print("Program execution completed!")
```

### **custom exception**

```python
production = int(input("Enter production count : "))
if production < 0:
    raise ValueError("Out of stock")
else:
    print("Product is available")
```

```python
class AgeError(Exception):
    pass 
age = int(input("Enter age : "))
if age < 18:
    raise AgeError("Age must be 18 or above")
else: print("Eligible")
```
